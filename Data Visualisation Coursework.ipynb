### Topic definitions:


**Mental wellbeing**: "Mental wellbeing derives from psychological functioning, which includes the ability to develop and maintain mutually beneficial relationships, and from levels of happiness and contentment with life, usually measured as life satisfaction. Psychological functioning includes the ability to maintain a sense of autonomy, agency, self-acceptance, self-esteem, personal growth and purpose in life. Mental wellbeing is more than the outcome of treating or preventing mental illness." (Warwick-Edinburgh Mental Wellbeing Scale, 2006)

**Eudaimonic perspective**: An "Eudaimonic perspective of wellbeing focuses on psychological functioning, good relationships with others and self-realisation. This is the development of human potential that, when realised, results in positive functioning in life, and covers a wide range of cognitive aspects of mental health." (Warwick-Edinburgh Mental Wellbeing Scale, 2006)

**Warwick edinburgh mental wellbeing scale (WEMWBS)**:

- Overview
: "The Warwick-Edinburgh Mental Wellbeing Scales were developed to enable the measuring of mental wellbeing in the general population and the evaluation of projects, programmes and policies which aim to improve mental wellbeing… The items are all worded positively and cover both feeling and functioning aspects of mental wellbeing, thereby making the concept more accessible. The scale has been widely used nationally and internationally for monitoring, evaluating projects and programmes and investigating the determinants of mental wellbeing." 
---

- How it is used
: "WEMWBS is a 14-item scale covering subjective wellbeing and psychological functioning, in which all items are worded positively and address aspects of positive mental health. The scale is scored by summing the response to each item answered on a 1 to 5 Likert scale. The minimum scale score is 14 and the maximum is 70. WEMWBS was initially validated for use in the UK with those aged 16 and above, involving surveys in both student and general population samples, and focus groups. It has now been widely validated in different populations and languages other than English."
---
 
- Origins
: The Warwick–Edinburgh Mental Well-being Scale was funded by the Scottish Government National Programme for Improving Mental Health and Wellbeing, commissioned by NHS Health Scotland, developed by the University of Warwick and the University of Edinburgh, and is jointly owned by NHS Health Scotland, the University of Warwick and the University of Edinburgh. (Professor Stewart-Brown and Dr Janmohamed, 2015)

```python
#Cleaning and preparing my three dataframes for analysis

# Import libraries 

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

When cleaning the data, specically, handling blank space and metadata, I initally tried using pandas bulit-in functions such as `dropna` but this removed a lot of key data or rearranged the dataset. Extra information addded by the editor meant a lot of repeated cleaning steps were needed. To improve efficiency I created a function called `remove_na_values()` to remove all types of metadata across each of my required datasets. 

`remove_na_values()` allows me to the repeated process of removing the inital blank spaces and colour graphics and focus on the key values. This is acheieved by removing the extra details, above, below and to the side of the orignal excell spreedsheets. 

Values such as is `negative_bottom` are counted and then entered into the funtion.

More specifc and tailored data manipulation will be done afterwards.

```python

## User-defined function for inital data cleaning, refer back to for keyword arguments
# we use slicing with the given arguments to removed various repeated unnecessary values. 

def remove_na_values(dataset, first_rows, last_column, middle_columns, negative_bottom ):
    
    # deleting the first set of rows of grpahics added by the editor to show overall change
    dataset.drop(index = dataset.index[0:first_rows], inplace = True)
    
    # removing all unnecessary columns from last_column onwards
    dataset.drop(columns=dataset.columns[last_column:], inplace = True)
    
    # check if there are middle columns to delete, if so remove them
    if middle_columns != None:
    # deleting formatting rows of blank space inbetween values
        dataset.drop(columns=middle_columns, inplace = True)
        
    
    # deleting the rows with editors notes at the bottom of the dataset
    dataset.drop(index=dataset.index[negative_bottom:], inplace=True)
    
    # Resetting the index back to zero to be more useful
    dataset.reset_index(drop=True,inplace=True)
    
    # Making the header to the likert scale options instead of 'Unnamed'
    dataset.columns = dataset.iloc[0]
    
    # Correcting a subscript error
    dataset.replace('2009-103', '2009-10', inplace = True)
    
    return dataset
```

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
